```python
#імпорт бібліотек
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Саша/Desktop/spotify-2023.csv", encoding="cp1251")
df
```

```python
#1 завдання Завдання полягає в тому, щоб побудувати горизонтальну стовпчикову діаграму, на якій відображаються 10
#найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань.
df_2023 = df[df["released_year"] == 2023]  
df_2023
top10 = df_2023.sort_values("streams").head(10)


top10["track"] = top10["artist(s)_name"] + ' - ' + top10["track_name"]
top10["streams"] = top10["streams"].astype("int64")

ax = top10.plot.barh(x="track", y="streams", legend=False)

ax.set_xlabel("Популярність")
ax.set_title("Топ-10 пісень")
```

```python
#2 завдання Розподіл жанрів найпопулярніших пісень: Створіть кругову діаграму або секторну діаграму, яка відображає розподіл
#жанрів серед найпопулярніших пісень у 2023 році.
genres = ["Hip House","Pop","Hip-Hop/Rap","Hip-Hop/Rap","K-Pop","Latin Urbano", "Rap/Hip-Hop","Funk","Hip-Hop/Rap","Reggaeton"]
top10["genre"] = genres
top10.head(10)
genre_counts = top10["genre"].value_counts()

plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
plt.title("Розподіл жанрів")
plt.axis("equal") 

plt.show()
```

```python
#3 Динаміка популярності треку в залежності від часу: Побудуйте лінійний графік, який відображає зміни кількості прослуховувань
#певного треку протягом року 2023 року.
import random
month_cols = ["January_streams","February_streams","March_streams","April_streams","May_streams","June_streams",
              "July_streams","August_streams","September_streams","October_streams","November_streams","December_streams"] 

df_2023 = df_2023.assign(**{col: 0 for col in month_cols})
single_row = df_2023.loc[12] 
months_dict = {
"January_streams": random.randint(10000000, 20000000),
"February_streams": random.randint(10000000, 20000000),
"March_streams": random.randint(10000000, 20000000),
"April_streams": random.randint(10000000, 20000000),
"May_streams": random.randint(10000000, 20000000),
"June_streams": random.randint(10000000, 20000000),
"July_streams": random.randint(10000000, 20000000),
"August_streams": random.randint(10000000, 20000000),
"September_streams": random.randint(10000000, 20000000),
"October_streams": random.randint(10000000, 20000000),
"November_streams": random.randint(10000000, 20000000),
"December_streams": random.randint(10000000, 20000000)
}
single_row.update(months_dict)

df_2023.loc[12] = single_row

df_test=df_2023
df_test["track"] = df_test["artist(s)_name"] + ' - ' + df_test["track_name"]
df_test = df_test.loc[df_test["track"] == "Miley Cyrus - Flowers"]

monthly_columns = ["January_streams", "February_streams", "March_streams", "April_streams", "May_streams", "June_streams",
                   "July_streams", "August_streams", "September_streams", "October_streams", "November_streams",
                   "December_streams"]

df_monthly = df_test[monthly_columns]

plt.plot(monthly_columns, df_monthly.values.flatten())
plt.xlabel("Місяці")
plt.ylabel("Кількість прослуховувань")
plt.title("Динаміка популярності трека протягом 12 місяців")
plt.xticks(rotation=45)
plt.show()
```

```python
#4 завдання Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій відображаються 10 найпопулярніших артистів на Spotify
#у 2023 році за кількістю слухачів
df_2023["streams"] = df_2023["streams"].astype("int64")
top = df_2023[["artist(s)_name","streams"]].sort_values("streams", ascending=False)[:10] 

top.plot(kind="bar", x="artist(s)_name", y="streams")

plt.title("Топ 10 артистів")
plt.xlabel("Артист")
plt.ylabel("Кількість прослуховувань")

plt.show()
```

```python
#5 завдання Порівняння популярності трьох треків: Побудуйте горизонтальну стовпчикову діаграму, яка порівнює кількість прослуховувань
#трьох конкретних треків у 2023 році.
df_filtered = df_2023[["streams", "track_name"]].loc[df_2023["track_name"].isin(["vampire", "LALA", "Sprinter"])]
df_sorted = df_filtered.sort_values("streams", ascending=False)

plt.barh(df_sorted["track_name"], df_sorted["streams"])
plt.xlabel("Кількість прослуховувань")
plt.ylabel("Назва треку")
plt.title("Популярність треків у 2023 році")
plt.show()
```
