```python
#5. Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та виконайте на них поелементні операції додавання, віднімання та множення

arr1 = np.random.randint(1, 21, size=6) 
arr2 = np.random.randint(1, 21, size=6)

print(arr1)
print(arr2)

#математичні операції
print(arr1 + arr2) 
print(arr1 - arr2)
print(arr1 * arr2)
```

```python
#6. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
arr1 = np.random.randint(1, 6, size=(3,2))
arr2 = np.random.randint(1, 6, size=(2,4))

print(arr1) 
print(arr2)

result = np.dot(arr1, arr2)
print(result)
```

```python
#7. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

matrix = np.random.rand(3,3)

print(matrix)

print("Транспонована матриця") 
print(matrix.T)
```

```python
#8. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.

arr1 = np.random.randint(1,11,(3,3)) 
arr2 = np.random.randint(1,11,(3,3))

#1 матриця
print(arr1)

#2 матриця 
print(arr2)

result = arr1 - arr2

print("Різниця")
print(result)
```

```python
#9. Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток

arr1 = np.random.randint(-5, 6, size=(3,3))
arr2 = np.random.randint(-5, 6, size=(3,3))

#1 матриця
print(arr1) 

#2 матриця

print(arr2)

result = arr1 * arr2

print(result)
```

```python
#10. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.

arr = np.random.uniform(-1, 1, size=(3,3))

#матриця
print(arr)

result = arr.sum(axis=1)

print(result)
```

```python
import pandas as pd

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96' 

tables = pd.read_html(url)

table = tables[20] 

table
```

```python
#7. Видалити дані для всієї країни з таблиці в останньому рядку таблиці.

table1= table.drop(table.index[-1])
table1
```

```python
#8. Замінити пропущені дані у стовпцях середніми значеннями за стовпчиком  (метод fillna)
table11= table1.fillna(table1.mean())
table11
```

```python
#9. Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні


mean_2014 = table11["2014"].mean()
df_2014 = table11["2014"]
high_values = df_2014[df_2014 > mean_2014]
high_indexes = high_values.index

regions = table11.loc[high_indexes, "регіон"].tolist()

print(regions)
```

```python
#10. В якому регіоні була найвища народжуваність у 2010 та 2012 роках.
#В таблиці 2010 року нема
print(table11.loc[table11["2012"].nlargest(1).index, "регіон"]) 
print(table11.loc[table11["2008"].nlargest(1).index, "регіон"])

```

```python
#11. Побудуйте стовпчасту народжуваності за регіонами у 2014 році

#імпорт matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 8)) 

ax.bar(x=table11["регіон"], height=table11["2014"]) 

ax.set_xticks(range(len(table11["регіон"])))
ax.set_xticklabels(table11["регіон"], rotation=90)

ax.set_xlabel("Регіон")
ax.set_ylabel("Народжуваність")
ax.set_title("Народжуваність по регіонам в 2014")  

fig.tight_layout()
plt.show()
```
