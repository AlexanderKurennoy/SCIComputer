```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("C:/Users/Саша/Desktop/shopping_trends.csv", encoding="cp1251")
```

```python
#підготовка датафрейму
print(df.shape)
print(df.isnull().sum())
print(df.dtypes)
df = df.rename(columns=lambda x: x.lower().replace(" ", "_"))
print(df.columns)
df.dtypes
```

```python
#Гістограма віку (age):

sns.displot(df["age"])


plt.title("Розподіл віку")
plt.xlabel("Вік")

plt.show()
```

```python
#2 Порівняння розподілу покупок за статтею (item_purchased)
purchases = df["item_purchased"].value_counts()
sns.barplot(x=purchases.index, y=purchases)
plt.rc("figure", figsize=(30, 6)) 
plt.title("Кількість покупок за товарами")
plt.ylabel("Кількість покупок")
```

```python
#3 Залежність між віком та сумою покупок (purchase_amount)
sns.scatterplot(x="age", y="purchase_amount_(usd)", data=df)
plt.rc("figure", figsize=(15, 12))
plt.title("Залежність віку та суми покупки")
plt.xlabel("Вік")
plt.ylabel("Сума покупки")

plt.show()
```

```python
#4 Графік розподілу оцінок товарів (review_rating)
sns.displot(df["review_rating"])
plt.title("Розподіл оцінок товарів")
plt.xlabel("Оцінка")
plt.show()
```

```python
#5 Категоріальна діаграма для статі (gender)
sns.catplot(x="gender", kind="count", data=df)
plt.title("Розподіл покупок за статтю")
plt.xlabel('Стать')

plt.show()
```

```python
#6 Залежність між рейтингом і сумою покупок (review_rating і purchase_amount)
sns.scatterplot(x="review_rating", y="purchase_amount_(usd)", data=df)
plt.rc("figure", figsize=(15, 12))
plt.title("Залежність рейтингу та суми покупки")
plt.xlabel("Рейтинг")
plt.ylabel("Сума покупки")


plt.show()
```

```python
#7 Сезонність покупок (season)
seasons = df["season"].value_counts()
sns.barplot(x=seasons.index, y=seasons)

plt.title("Кількість покупок за сезонами")
plt.ylabel("Кількість покупок")
plt.xlabel("Сезон")

plt.show()
```

```python
#8 Графік розподілу розмірів одягу (size)
sns.countplot(x="size", data=df)
plt.show()
```

```python
#9 Кількість покупок відносно типу оплати (payment_method)
payment_counts = df['payment_method'].value_counts()
sns.barplot(x=payment_counts.index, y=payment_counts)
plt.show()
```

```python
# 10 Залежність між кількістю попередніх покупок (previous_purchases) і сумою поточної покупки (purchase_amount)
sns.scatterplot(x="previous_purchases", y="purchase_amount_(usd)", data=df)
plt.rc("figure", figsize=(15, 12))
plt.title("Зв`язок між попередніми покупками та сумою поточної")
plt.xlabel("Кількість попередніх покупок")
plt.ylabel("Сума поточної покупки")
plt.show()
```
